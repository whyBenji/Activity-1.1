```python
def ordinal_encode(df, column_order_map):
    encoder = OrdinalEncoder()
    for col, order in column_order_map.items():
        encoder.set_params(categories=[order])  # Set the custom order
        df[col] = encoder.fit_transform(df[[col]])  # Transform the column
    return df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold


catVar = make_column_selector(dtype_include=['object'])
#catVar = ['Q6']
# Cross-validation for the model
def modelPipeline(modelName, X=X_train, y=y_train):
    # Define the column transformer to standardize and encode categorical features
    ct = ColumnTransformer(
        transformers=[
            ('standardize', StandardScaler(), ['Q2']),  # Adjust this column as needed
            ('encode', OneHotEncoder(drop='first', sparse_output=False), catVar),  # Specify categorical columns,
            #('encode_ordinal', FunctionTransformer(ordinal_encode, kw_args={'column_order_map': custom_order}), ['education_level', 'income_bracket']),

            
        ], 
        remainder='passthrough'  # Keep other columns as they are
    )
    
    # Create the full pipeline with column transformation and model fitting
    pipeline = Pipeline(
        steps=[
            ('columnTransform', ct),
            ('Model', modelName)
        ]
    )
    
    # Perform cross-validation (cv=10 for 10-fold cross-validation)
    cv = StratifiedKFold(n_splits=10)
    cross_val_scores = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')
    
    # Calculate and print the average cross-validation score
    mean_cv_score = cross_val_scores.mean()
    print("Cross-validation Accuracy:", mean_cv_score)

    return pipeline, mean_cv_score

# Hyperparameter tuning with cross-validation
def tunning(model, X, y, paramGrid=None):
    if paramGrid is None:
        paramGrid = {
        'Model__penalty': ['l1', 'l2'],
        'Model__C': [0.001, 0.01, 0.1, 1, 10, 100],  
        'Model__solver': ['liblinear', 'newton-cg', 'saga', 'lbfgs'], 
        'Model__multi_class': ['ovr', 'multinomial'],    # Multi-class strategy ('ovr' or 'multinomial')
        
}
#         paramGrid = {
#         'Model__penalty': ['l1', 'l2', 'elasticnet'],    # L1, L2, or ElasticNet regularization
#         'Model__C': [0.001, 0.01, 0.1, 1, 10, 100],     # Regularization strength (inverse)
#         'Model__solver': ['liblinear', 'newton-cg', 'saga', 'lbfgs'],  # Solvers for optimization
#         'Model__max_iter': [100, 200, 300],              # Number of iterations for convergence
#         'Model__class_weight': [None, 'balanced'],        # Adjust for class imbalance
#         'Model__tol': [1e-4, 1e-3, 1e-2],                # Tolerance for optimization
#         'Model__multi_class': ['ovr', 'multinomial'],    # Multi-class strategy ('ovr' or 'multinomial')
#         'Model__warm_start': [True, False],   # Whether to reuse the solution of the previous call
        
# }

    # Create a pipeline with column transformation and model fitting for grid search
    ct = ColumnTransformer(
        transformers=[
            ('standardize', StandardScaler(), ['Q2']),  # Adjust this column as needed
            ('encode', OneHotEncoder(sparse_output=False), make_column_selector(dtype_include=['object'])),  # Specify categorical columns
        ], 
        remainder='passthrough'
    )
    
    pipeline = Pipeline(
        steps=[
            ('columnTransform', ct),
            ('Model', model)
        ]
    )
    
    # Perform grid search with cross-validation (cv=10 for 10-fold cross-validation)
    gridSearch = GridSearchCV(pipeline, param_grid=paramGrid, scoring='accuracy', cv=10)
    gridSearch.fit(X, y)
    
    return gridSearch.best_params_

# Example of using the pipeline and tuning function
lr = LogisticRegression(multi_class='ovr')
model, accuracy = modelPipeline(lr, X, y)  # Now uses cross-validation
best_params = tunning(lr, X, y)
print("Best parameters:", best_params)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline

def modelPipeline(modelName, X=X_train, y=y_train):
    # Define the column transformer to standardize and encode categorical features
    ct = ColumnTransformer(
        transformers=[
            #('standardize', StandardScaler(), ['Q2']),  # Adjust this column as needed
            ('encode', OneHotEncoder(drop='first', sparse_output=False), make_column_selector(dtype_include=['object'])),  # Specify categorical columns
        ], 
        remainder='passthrough'  # Keep other columns as they are
    )
    
    # Create the full pipeline with column transformation and model fitting
    pipeline = Pipeline(
        steps=[
            ('columnTransform', ct),
            ('Model', modelName)
        ]
    )
    
    # Perform cross-validation (cv=10 for 10-fold cross-validation)
    cv = StratifiedKFold(n_splits=10)
    cross_val_scores = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')
    
    # Calculate and print the average cross-validation score
    mean_cv_score = cross_val_scores.mean()
    print("Cross-validation Accuracy:", mean_cv_score)

    return pipeline, mean_cv_score

def tunning(model, X=X_train, y=y_train, paramGrid=None):
    if paramGrid is None:
        paramGrid = {
    'Model__criterion': ['gini', 'entropy'],  # Splitting criteria
    'Model__max_depth': [3, 5, 10, 15, 20, None],  # Depth of the tree (None means no limit)
    'Model__min_samples_split': [2, 5, 10, 20],  # Minimum samples required to split an internal node
    'Model__min_samples_leaf': [1, 2, 4, 10],  # Minimum samples required to be at a leaf node
    'Model__max_features': ['sqrt', 'log2', None],  # Number of features to consider for the best split
    'Model__splitter': ['best', 'random'],  # Strategy used to split at each node
    'Model__class_weight': [None, 'balanced'],  # Weights associated with classes
    'Model__min_impurity_decrease': [0.0, 0.01, 0.05],  # Minimum impurity decrease required for a split
}


    # Create a pipeline with column transformation and model fitting for grid search
    categorical_columns = make_column_selector(dtype_include=['object'])
    
    ct = ColumnTransformer(
        transformers=[
            #('standardize', StandardScaler(), ['Q2']),  # Adjust this column as needed
            ('encode', OneHotEncoder(sparse_output=False), categorical_columns),  # Specify categorical columns
        ], 
        remainder='passthrough'
    )
    
    pipeline = Pipeline(
        steps=[
            ('columnTransform', ct),
            ('Model', model)
        ]
    )
    
    # Perform grid search with cross-validation (cv=10 for 10-fold cross-validation)
    gridSearch = GridSearchCV(pipeline, param_grid=paramGrid, scoring='accuracy', cv=10)
    gridSearch.fit(X, y)
    
    return gridSearch.best_params_

# Example of using the pipeline and tuning function
dt = DecisionTreeClassifier()
model, accuracy = modelPipeline(dt, X, y)  # Now uses cross-validation
best_params = tunning(dt, X, y)
print("Best parameters:", best_params)
```
